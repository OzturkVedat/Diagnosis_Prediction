```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split

df= pd.read_csv("hearth_diseases_dataset.csv")                # first things first, data-preprocessing
df.isna().sum()                     # checking for missing values

x = df.drop(['target','patientid'], axis=1)  # Features
y = df['target']               # Target variable

# 80-20 split for train-test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training set shape:", x_train.shape, y_train.shape)     
print("Testing set shape:", x_test.shape, y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.linear_model import LogisticRegression
        # training and tuning the Logistic Regression model
parameters = {
    'C' : [0.001, 0.01, 0.1, 1.0, 10, 100, 1000],
    'class_weight' : ['balanced'],
    'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
lr_model = LogisticRegression()

lr_cv = GridSearchCV(estimator=lr_model, param_grid=parameters, cv=10).fit(x_train, y_train)

print('Tuned hyper parameters : ', lr_cv.best_params_)

```

```python
import joblib       # for loading the model to an API

joblib.dump(lr_cv, 'path/to/your_model.pkl')        # edit your directory 
```
