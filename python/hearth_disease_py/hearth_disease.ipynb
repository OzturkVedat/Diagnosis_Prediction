```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split

df= pd.read_csv("hearth_diseases_dataset.csv")                # first things first, data-preprocessing
df.isna().sum()                     # checking for missing values

y = df['target']               # Target variable
x = df.drop(['target','patientid'], axis=1)  # Features

print("Feature names:", x.columns.tolist())
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.linear_model import LogisticRegression
        # training and tuning the Logistic Regression model
parameters = {
    'C' : [0.001, 0.01, 0.1, 1.0, 10, 100, 1000],
    'class_weight' : ['balanced'],
    'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
lr_model = LogisticRegression()

lr_cv = GridSearchCV(estimator=lr_model, param_grid=parameters, cv=10).fit(x, y)

print('Tuned hyper parameters : ', lr_cv.best_params_)

```

```python
import joblib       # for loading the model to an API

joblib.dump(lr_cv, 'logRes_model.pkl')        # edit your directory 
```
